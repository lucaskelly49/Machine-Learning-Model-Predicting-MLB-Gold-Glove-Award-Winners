```python
# Standard SS fielding statitics from Fangraphs

dfSS = pd.read_csv('FanGraphs Leaderboard (1).csv')
dfSS = dfSS.drop(['rTS', 'rCERA', 'rSB', 'rSZ', 'rARM', 'FSR', 'FRM', 'ARM'], axis = 1)
dfSS.head()
```

```python
# Advanced SS fielding statistics from Fangraphs

dfSSA = pd.read_csv('FanGraphs Leaderboard (2).csv')
dfSSA = dfSSA.drop(['Name', 'Team', 'Pos', 'Inn', 'DPT', 'DPF', 'Scp', 'SB', 'CS', 'PB', 'WP', 'FE', 'TE', 'DPS', 'TZ'], axis = 1)
dfSSA.head()
```

```python
dfSS_clean = pd.merge(dfSS, dfSSA, on=['Season', 'playerid'])
```

```python
# Advanced 2B fielding statistics from Fangraphs

df2B = pd.read_csv('FanGraphs Leaderboard (2BA).csv')
df2B = df2B.drop(['rTS', 'rCERA', 'rSB', 'rSZ', 'rARM', 'FSR', 'FRM', 'ARM'], axis = 1)
df2B.head()
```

```python
# Standard 2B fielding statistics from Fangraphs

df2BA = pd.read_csv('FanGraphs Leaderboard (2B).csv')
df2BA = df2BA.drop(['Name', 'Team', 'Pos', 'Inn', 'DPT', 'DPF', 'Scp', 'SB', 'CS', 'PB', 'WP', 'FE', 'TE', 'DPS', 'TZ'], axis = 1)
df2BA.head()
```
